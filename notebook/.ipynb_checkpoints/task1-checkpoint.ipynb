```python
# =========================================================
# EthioMart Telegram Scraper ‚Äì Task 1 (text + images)
# =========================================================
# Prereqs: you have already run
#   from telethon import TelegramClient
#   client = TelegramClient('ethio_scraper_session', api_id, api_hash)
#   await client.start(phone=phone)

import os, re, unicodedata, pandas as pd, nest_asyncio
nest_asyncio.apply()   # allow nested awaits inside Jupyter

# 1Ô∏è‚É£  Select any 5 channels to scrape
CHANNELS = [
    "@ZemenExpress", "@nevacomputer", "@meneshayeofficial", "@ethio_brand_collection", "@Leyueqa",
    "@sinayelj", "@Shewabrand", "@helloomarketethiopia", "@modernshoppingcenter", "@qnashcom",
    "@Fashiontera", "@kuruwear", "@gebeyaadama", "@MerttEka", "@forfreemarket", "@classybrands",
    "@marakibrand", "@aradabrand2", "@marakisat2", "@belaclassic", "@AwasMart"
]
channels_to_scrape = CHANNELS[:5]               # first 5 for demo
pd.DataFrame(channels_to_scrape, columns=["channel"]).to_excel(
    "channels_to_crawl.xlsx", index=False
)

# 2Ô∏è‚É£  Simple Amharic text cleaner
def clean_text(text: str) -> str:
    if not text:
        return ""
    text = unicodedata.normalize("NFC", str(text))
    text = text.replace("\n", " ").strip()
    text = re.sub(r"\s+", " ", text)
    tokens = re.findall(r"\w+", text, flags=re.UNICODE)
    return " ".join(tokens)

# 3Ô∏è‚É£  Async scraping coroutine
async def scrape_channels(client, channels, limit=1000, media_dir="photos"):
    os.makedirs(media_dir, exist_ok=True)
    rows = []
    for username in channels:
        try:
            entity = await client.get_entity(username)
        except Exception as e:
            print(f"‚ö†Ô∏è  Couldn't access {username}: {e}")
            continue
        title = getattr(entity, "title", str(entity))
        count = 0
        async for msg in client.iter_messages(entity, limit=limit):
            if not (msg.message or msg.media):
                continue
            media_path = None
            # download photo, if any
            if msg.media and getattr(msg.media, "photo", None):
                fname = f"{username.strip('@')}_{msg.id}.jpg"
                fpath = os.path.join(media_dir, fname)
                try:
                    await client.download_media(msg.media, fpath)
                    media_path = fpath
                except Exception as e:
                    print(f"   ‚è© Failed to download {fname}: {e}")
            rows.append(
                {
                    "channel_title": title,
                    "channel_handle": username,
                    "message_id": msg.id,
                    "date": msg.date,
                    "raw_text": (msg.message or "").strip(),
                    "clean_text": clean_text(msg.message),
                    "media_file": media_path,
                }
            )
            count += 1
        print(f"‚úÖ {username}: scraped {count} messages")
    return pd.DataFrame(rows)

# 4Ô∏è‚É£  Run the scraper and save output
df = await scrape_channels(client, channels_to_scrape, limit=1000)
df.to_csv("processed_telegram_messages.csv", index=False)
print(f"\nüíæ Saved {len(df)} rows to processed_telegram_messages.csv")
df.head()
```
